```python

def plot_graph(data,v,color = 'lightblue'):
    x = [i for i in range(1,len(data))]
    y = [data[i] for i in range(1,len(data))]
    #y = list(map(float, y))
    plt.plot(x,y,color=color,linewidth = 0.7)
    plt.xlabel('Episodes') # naming the x axis      
    plt.ylabel('Rewards') # naming the y axis
    plt.title('Episode vs Reward') 
```

```python
def plot_graph_2(data,v,color):
    x = [i for i in range(1,len(data))]
    y = [data[i] for i in range(1,len(data))]
    #y = list(map(float, y))
    plt.plot(x,y,label=v,color = color,linewidth = 3)
    plt.xlabel('Episodes') # naming the x axis      
    plt.ylabel('Rewards') # naming the y axis
    plt.title('Episode vs Reward') 
```

```python
def plot_graph_3(data,v,color):
    x = [i for i in range(1,len(data))]
    y = [data[i][1] for i in range(1,len(data))]
    y = list(map(float, y))
    plt.plot(x,y,label=v,color = color,linewidth = 3)
    plt.xlabel('Episodes') # naming the x axis      
    plt.ylabel('Rewards') # naming the y axis
    plt.title('Episode vs Reward') 
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame
import pandas as pd

averaged_plot = []
plt.figure(figsize=(8, 6))

    
f1 = 'ddpg_2singleClients.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')

    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        #plot_graph(d,'client ' + str(i),color = 'khaki')
        #avg = df.mean(axis=1)
        averaged_plot.append(d)
    
f1 = 'ddpg_avgGrad_seed.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')

    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'orange')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
f1 = 'ddpg_Concurrent_seed13.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')
    
    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'palegreen')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)

plot_graph_2(averaged_plot[0],'Single Client 1',color = 'dodgerblue')
plot_graph_2(averaged_plot[1],'Single Client 2',color = 'blue')
plot_graph_2(averaged_plot[2],'Fed Avg ',color = 'orange')
plot_graph_2(averaged_plot[3],'Concurrent Avg',color = 'green')

'''
f1 = 'fed_pytorch_concurrentRL_after4epsiodes.csv'    
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    plot_graph_3(data,'Concurrent set 2',color = 'red')
'''

plt.legend(loc='best')    
plt.show()
```
