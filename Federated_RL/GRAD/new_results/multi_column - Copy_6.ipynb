```python

def plot_graph(data,v,color = 'lightblue'):
    x = [i for i in range(1,len(data))]
    y = [data[i] for i in range(1,len(data))]
    #y = list(map(float, y))
    plt.plot(x,y,color=color,linewidth = 0.3)
    plt.xlabel('Episodes') # naming the x axis      
    plt.ylabel('Rewards') # naming the y axis
    plt.title('Episode vs Reward') 
```

```python
def plot_graph_2(data,v,color):
    x = [i for i in range(1,len(data))]
    y = [data[i] for i in range(1,len(data))]
    #y = list(map(float, y))
    plt.plot(x,y,label=v,color = color,linewidth = 3)
    plt.xlabel('Episodes') # naming the x axis      
    plt.ylabel('Rewards') # naming the y axis
    plt.title('Episode vs Reward') 
```

```python
def plot_graph_3(data,v,color):
    x = [i for i in range(1,len(data))]
    y = [data[i][1] for i in range(1,len(data))]
    y = list(map(float, y))
    plt.plot(x,y,label=v,color = color,linewidth = 3)
    plt.xlabel('Episodes') # naming the x axis      
    plt.ylabel('Rewards') # naming the y axis
    plt.title('Episode vs Reward') 
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame
import pandas as pd

plt.figure(figsize=(11, 8))
averaged_plot = []
    
f1 = 'GradAvg_seed1.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')

    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'cornflowerblue')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
f1 = 'GradAvg_seed2.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')

    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'mediumpurple')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
f1 = 'BestScore_seed1.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')
    
    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'lightcoral')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
f1 = 'BestScore_seed2.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')
    
    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'lightsalmon')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
f1 = 'WeightAvg_seed1.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')
    
    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'yellow')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
f1 = 'WeightAvg_seed2.csv'
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    df = DataFrame(data)
    df = df.drop(df.columns[0], axis=1)
    df = df.apply(pd.to_numeric, errors='coerce')
    
    for i in range(0,len(df.columns)):
        d = df.iloc[:,i]
        plot_graph(d,'client ' + str(i),color = 'forestgreen')
    avg = df.mean(axis=1)
    averaged_plot.append(avg)
    
    
'''   
f1 = 'fed_pytorch_concurrentRL_after4epsiodes.csv'    
with open(f1, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    plot_graph_3(data,'Concurrent set 1',color = 'blue')
#plot_graph_2(averaged_plot[0],'Target update after 4 set 1',color = 'dodgerblue')
'''
plot_graph_2(averaged_plot[0],'Fed Avg set 1',color = 'indigo')
plot_graph_2(averaged_plot[1],'Fed Avg set 2',color = 'blue')
plot_graph_2(averaged_plot[2],'Best Score set 1',color = 'red')
plot_graph_2(averaged_plot[3],'Best Score set 2',color = 'orange')
plot_graph_2(averaged_plot[4],'Weight Avg set 1',color = 'lime')
plot_graph_2(averaged_plot[5],'Weight Avg set 2',color = 'darkgreen')

plt.legend(loc='best')    
plt.show()
```
